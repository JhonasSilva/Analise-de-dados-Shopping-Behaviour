#  - Limpeza e tratamento dos dados
Vamos realizar o tratamento dos dados para corrigir inconsistências,
pradonizar informações, tratar valores ausentes e preparar o dataset
para futuras analises.

```python
import pandas as pd

df = pd.read_csv(r"C:\Analise de dados_Shopping_Behaviour_Dataset\Dados\shopping_behavior_updated.csv")

df.head()
```

```python
df.isnull().sum()
```

# Colunas duplicadas:
- Não foi encontrado valores duplicados no dataset

```python
df.dtypes
```

```python
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].astype(float)
```

```python
df['Purchase Amount (USD)'].dtypes
```

```python
# Esse comando ele pradoniza os texto de cada variavel dentro da lista "colunas_texto"
colunas_texto = ['Gender', 'Category', 'Season', 'Location']

for col in colunas_texto:
    df[col] = df[col].str.strip().str.title()
```

# Conversão de dados
- Foi indentificado que a coluna "Purchase Amount (USD)" estava com o tipo inteiro, foi convertido para flooat

```python
df.to_csv(r'C:\Analise de dados_Shopping_Behaviour_Dataset\Dados\dataset_tratado.csv', index=False) # Salvar o csv na pasta dados
```
