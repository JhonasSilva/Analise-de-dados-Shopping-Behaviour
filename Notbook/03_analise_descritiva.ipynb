# 3 - Analise descritiva
Nessa etapa vai ser realizada analises descritiva com intuito de entender o 
comportamento geral das vendas, perfil dos clientes e a distribuição dos principais indicadores de negócio.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\JhoooN4Zz\OneDrive\Desktop\Meus breguesso de programação\Projetos\Analise de dados_Shopping_Behaviour_Dataset\Dados\dataset_tratado.csv")
df.head()
```

### Faturamento total:

```python
faturamento_total = df['Purchase Amount (USD)'].sum()
print(f'(USD) {faturamento_total:,.2f}')
```

### Media do faturamento:

```python
media_faturamento = df['Purchase Amount (USD)'].mean()
print(f'(USD) {media_faturamento:,.2f}')
```

- Observação:

-O faturamento total foi feito com a soma dos valores da coluna "Purchase Amount (USD)",
resutando no total de 233.081,00.

-A media foi (USD) 59.76, indicando o valor gasto por compra de consumo dos cliente.

## Vendas por categorias:

```python
vendas_categoria = df.groupby('Category')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
vendas_categoria
```

```python
# Plot do grafico da vandas por categoria:
plt.figure(figsize=(8,5))
vendas_categoria.plot(kind='bar', color='b')

# Colocando o valor em cada barra
for i, valor in enumerate(vendas_categoria):
    plt.text(i, valor + 0.1, str(f'{valor:,.2f}'), ha='center', fontsize=10)

# Descrição do grafico
plt.title('Faturamento por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Faturamento (USD)')
plt.xticks(rotation=1)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Salvar a imagem do grafico na pasta imagens.
plt.savefig(r"C:\Users\JhoooN4Zz\OneDrive\Desktop\Meus breguesso de programação\Projetos\Analise de dados_Shopping_Behaviour_Dataset\Imagens/grafico_categoria.png", dpi=300, bbox_inches="tight")
```

- Observação:

-Podemos ver que a categoria 'Clothing' domina no rank, pertencendo uma fatia maior no faturamento da empresa, com os seus (USD) 104.264,00 colocando ela em primeiro lugar. E a categoria que menos vendeu foi a Outerwear (USD) 18.524,00 pertencendo a fatia menor.

## Vendas por gênero:

```python
vendas_genero = df['Gender'].value_counts()
vendas_genero
```

```python
plt.figure(figsize=(8,5))
vendas_genero.plot(kind='bar', color='r')

# Colocando o valor em cada barra
for i, valor in enumerate(vendas_genero):
    plt.text(i, valor + 0.5, str(f'{valor:,.2f}'), ha='center', fontsize=10)

# Descrição do grafico
plt.title('Distribuição de Compras por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Quantidade de Compras')
plt.xticks(rotation=1)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Salvar a imagem do grafico na pasta imagens.
plt.savefig(r"C:\Users\JhoooN4Zz\OneDrive\Desktop\Meus breguesso de programação\Projetos\Analise de dados_Shopping_Behaviour_Dataset\Imagens/grafico_genero.png", dpi=300, bbox_inches="tight")
```

## Ticket medio por gênero:

```python
ticket_genero = df.groupby('Gender')['Purchase Amount (USD)'].mean()
ticket_genero
```

- Obsevação:

 -O sexo masculino dispara na quantidade de total vendidos comparado com o sexo feminino, no total de 2.652 para masculino e 1.248 para feminino, indicando que o publimo masculino teve mais interesse nas compras da emprensa.

 -A media por sexo foi de (USD) 60,00 por compra de ambos os sexo, mostrando que a media de compra por cliente é igual

### VENDAS POR ESTAÇÃO DO ANO:

```python
vendas_estacao = df.groupby('Season')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
vendas_estacao
```

```python
plt.figure(figsize=(8,5))
vendas_estacao.plot(kind='bar', color='orange')

# Colocando o valor em cada barra
for i, value in enumerate(vendas_estacao):
    plt.text(i, value + 0.5, str(value), ha='center', fontsize=10)

# Descrição do grafico
plt.title('Faturamento por Estação')
plt.xlabel('Estação')
plt.ylabel('Faturamento (USD)')
plt.xticks(rotation=1)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Salvar a imagem do grafico na pasta imagens.
plt.savefig(r"C:\Users\JhoooN4Zz\OneDrive\Desktop\Meus breguesso de programação\Projetos\Analise de dados_Shopping_Behaviour_Dataset\Imagens/grafico_estação.png", dpi=300, bbox_inches="tight")
```

- Observação:

Outono foi a que mais faturou, com os seus (USD) 60.018, representando a estação do ano que teve maior lucro a empresa, e as outras estação teve totais bem parecido mostrando que a empresa teve seu pico de vendas estavel durante o ano.

## Media da Avaliação:

```python
avaliacao_media = df['Review Rating'].mean()
print(f'{avaliacao_media:.2f}')
```

- Observação:

-A avaliação média dos produtos é de 3,7, o que indica um nível de satisfação
moderadamente positivo por parte dos clientes. Apesar de a maioria das experiências
ser favorável, esse valor sugere que ainda existem pontos de melhoria na qualidade
dos produtos ou no serviço oferecido.

### Conclusão:

- Com a análise descritiva, observou-se que a empresa obteve um faturamento total de
USD 233.081,00, com um ticket médio de USD 59,76 por cliente. A categoria que mais
contribuiu para esse resultado foi Clothing, com um faturamento de USD 104.264,00,
representando aproximadamente 50% do faturamento total.

- Em relação ao perfil dos clientes, o sexo masculino apresentou maior participação
no número de compras, indicando que a maior parte do público é composta por homens.
Entretanto, o ticket médio por sexo mostrou-se equivalente entre os gêneros.

- Quanto à sazonalidade, a estação do outono foi a que apresentou o maior volume de
vendas, sendo o período de maior lucratividade para a empresa.

- A avaliação média dos produtos foi de 3,7, o que indica um nível de satisfação
moderadamente positivo por parte dos clientes, embora ainda exista margem para
melhorias na experiência oferecida.

### Principais insights extraídos:
